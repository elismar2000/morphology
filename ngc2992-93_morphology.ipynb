### To calculate the scale parameters, I used the following simple function to transform from pixels to physical units

```python
import numpy as np
def px2kpc(q, ps=1.0, distance=39):
    ps *= (1/3600) * np.pi/180 #from arcsecs to radians
    distance *= 1e+3 #distance from Mpc to Kpc
    q *= ps * distance
    return q
```

Using the results from the photometric fit, we have the following scale lengths for the stellar disk and bulge of the galaxies:

```python
#NGC2992
Re_bulge_2992 = 3.77
Rs_disk_2992 = 11.48

#NGC2993
Re_bulge_2993 = 2.03
Rs_disk_2993 = 4.17

#Transforming to physical units:
#NGC2992
Re_bulge_2992 = px2kpc(q=Re_bulge_2992)
Rs_disk_2992 = px2kpc(q=Rs_disk_2992)

#NGC2993
Re_bulge_2993 = px2kpc(q=Re_bulge_2993)
Rs_disk_2993 = px2kpc(q=Rs_disk_2993)

#From eqn 38 of Hernquist (1989), we can transform the bulge from Re (effective radius) to a (Hernquist scale parameter)
#This is because MakeNewDisk employs a Hernquist 1990 profile to build the bulge

a_bulge_2992 = Re_bulge_2992 / 1.8153
a_bulge_2993 = Re_bulge_2993 / 1.8153


print("NGC2992 bulge scale parameter = {:.3f} kpc".format(a_bulge_2992))
print("NGC2992 stellar disk scale length = {:.3f} kpc".format(Rs_disk_2992))

print("NGC2993 bulge scale parameter = {:.3f} kpc".format(a_bulge_2993))
print("NGC2993 stellar disk scale length = {:.3f} kpc".format(Rs_disk_2993))
```

```python
#From this, we conclude that

BulgeSize_2992 = a_bulge_2992 / Rs_disk_2992
BulgeSize_2993 = a_bulge_2993 / Rs_disk_2993

print("NGC2992 BulgeSize = {:.3f} kpc".format(BulgeSize_2992))
print("NGC2993 BulgeSize = {:.3f} kpc".format(BulgeSize_2993))
```
